```python
# utilizando SQLAlchemy e pandas.
# Importando a biblioteca SQLAlchemy para criar e manipular bancos de dados
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect
```

```python
# Criando um banco de dados SQLite em memória (temporário, desaparece ao fechar o kernel)
engine = create_engine('sqlite:///:memory:')
```

```python
import pandas as pd
```

```python
# Lendo o arquivo CSV com informações dos clientes
url = 'data/clientes_banco.csv'
```

```python
dados = pd.read_csv(url)
```

```python
# Visualizando as primeiras linhas do dataframe
dados.head()
```

```python
# Escrevendo os dados no banco de dados SQLite criado
# Cria uma tabela chamada 'clientes' com os dados do dataframe
dados.to_sql('clientes', engine, index=False)
```

```python
# Checando as tabelas existentes no banco de dados
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
# Realizando uma consulta SQL no banco de dados para selecionar apenas clientes empregados
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
```

```python
# Visualizando os dados dos empregados
empregados
```

```python
# Salvando o resultado da consulta em uma nova tabela chamada 'empregados'
empregados.to_sql('empregados', engine, index=False)
```

```python
# Lendo a tabela 'empregados' do banco de dados
pd.read_sql_table('empregados', engine)
```

```python
# Selecionando colunas específicas da tabela 'empregados'
pd.read_sql_table('empregados', engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

```python
# Atualizando dados em um banco de dados
# Primeiro, vamos ler todos os dados da tabela 'clientes'
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(query, engine)
```

```python
# Importando text para executar instruções SQL
from sqlalchemy import text
```

```python
# Exemplo de exclusão de um registro específico
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
with engine.connect() as conn:
  # Executa a consulta SQL
  result = conn.execute(text(query))
  # Confirma a alteração no banco de dados
  conn.commit()
```

```python
# Conferindo os dados após a exclusão
pd.read_sql_table('clientes', engine)
```

```python
# Exemplo de atualização de um registro específico
query = 'UPDATE clientes SET Grau_escolaridade ="Ensino superior" WHERE ID_Cliente =5008808'
with engine.connect() as conn:
  # Executa a consulta de atualização
  result = conn.execute(text(query))
  # Confirma a alteração
  conn.commit()
```

```python
# Conferindo os dados após a atualização
pd.read_sql_table('clientes', engine)
```
