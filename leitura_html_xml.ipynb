```python
import pandas as pd
```

```python
# Lê todas as tabelas de um arquivo HTML
dados_html = pd.read_html('data/filmes_wikipedia.html')
```

```python
# Mostra a lista de DataFrames extraídos do HTML
dados_html
```

```python
# Verifica o tipo do objeto retornado
type(dados_html)
```

```python
# Verifica quantas tabelas foram lidas do HTML
len(dados_html)
```

```python
# Seleciona a tabela que contém os 100 melhores filmes
top_filmes = dados_html[1]
```

```python
# Mostra a tabela dos 100 melhores filmes
top_filmes
```

```python
# ESCREVENDO ARQUIVOS HTML
# Exporta o DataFrame dos 100 melhores filmes para um arquivo HTML
top_filmes.to_html('data/top_filmes.html')
```

```python
# Lê novamente o arquivo gerado em HTML e mostra como tabela
pd.read_html('data/top_filmes.html')
```

```python
# ESCREVENDO ARQUIVOS CSV
# Exporta o DataFrame para CSV sem o índice numérico
top_filmes.to_csv('data/top_filmes_1998', index=False)
```
